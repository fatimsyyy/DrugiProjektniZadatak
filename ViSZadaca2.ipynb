## 1. Fitovanje distribucija i Chi-square Goodness of Fit test-Fatima Fulurija

## PREDVIĐANJE (prije analize)

### 1. Varijabla:
### Koju teorijsku distribuciju očekujemo i zašto?
Za varijablu 4. Your current GPA (grade point average) očekujemo teorijsku Normalnu distribuciju. Logika iza ovoga je da je akademski uspjeh na fakultetima odraz velikog broja različitih i nezavisnih faktora, kao što su uloženo vrijeme učenje, obaveze pored fakulteta, urođena sposobnost usvajanja gradiva, koncentracija na ispitima i slično. Na osnovu oblika distribucije iz prethodne zadaće, ocjene studenata obično prate klasičnu zvonastu krivu gdje je većina studenata grupisana oko središnjeg prosjeka (npr. $7.5 – 8.5$), dok su ekstremi znatno rjeđi.

###Da li očekujemo da će Chi-square test potvrditi ili odbaciti vaš izbor?
Očekujem da će Chi-square test potvrditi ovaj izbor. Budući da se radi o prirodnom akademskom indikatoru koji vizuelno balansira oko sredine, pretpostavljamo da će normalna distribucija pružiti statistički adekvatan model za opisivanje GPA.

### 2. Varijabla:
### Koju teorijsku distribuciju očekujemo i zašto?
Za varijablu 5. očekujemo teorijsku Log-normalnu distribuciju. Broj sati učenja je u realnosti obično desno asimetričan: većina studenata ulaže umjeren, sličan broj sati sedmično, dok postoji manja, specifična grupa studenata koji uče ekstremno mnogo (formirajući dugi desni rep). Za razliku od obične eksponencijalne distribucije, log-normalna distribucija znatno bolje opisuje početnu "grbu" na histogramu prije nego što započne strmi pad prema desnom repu ekstremnih vrijednosti.Da li očekujemo da će Chi-square test potvrditi ili odbaciti vaš izbor?
Očekujem da će Chi-square test potvrditi log-normalnu distribuciju. S obzirom na to da ovaj model matematički idealno toleriše i opisuje asimetriju i pozitivno nagnute repove, smatramo da će test potvrditi adekvatnost ovog fita.


```python

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Učitavanje podataka (prilagodi naziv fajla ako je potrebno)
df = pd.read_csv('Zadaca 1 - Dataset (1).csv', encoding='latin1')

kolona_gpa = "4. Your current GPA (grade point average)"
kolona_sati = "5. How many hours per week do you spend on university-related activities (lectures, studying, projects)?"

def fituj_i_testiraj(df_izvor, naziv_kolone, tip_distribucije="norm"):
    # Čišćenje podataka od NaN vrijednosti i pretvaranje u numerički tip
    podaci = pd.to_numeric(df_izvor[naziv_kolone], errors='coerce').dropna()

    # 1. Procjena parametara iz podataka (fitovanje)
    if tip_distribucije == "norm":
        params = stats.norm.fit(podaci)
        dist = stats.norm
        oznaka = "Normalna"
        Br_procijenjenih_param = 2  # mu i sigma
    elif tip_distribucije == "lognorm":
        params = stats.lognorm.fit(podaci, floc=0) # fiksiramo loc=0 radi stabilnosti sati učenja
        dist = stats.lognorm
        oznaka = "Log-Normalna"
        Br_procijenjenih_param = 2  # s i scale

    # 2. Vizuelno poređenje (Histogram + PDF kriva)
    plt.figure(figsize=(10, 5))
    sns.histplot(podaci, kde=False, stat="density", bins=10, label="Empirijski podaci (Anketa)", color="skyblue", edgecolor="0.2")

    # Generisanje tačaka za teorijsku krivu
    x = np.linspace(podaci.min(), podaci.max(), 200)
    pdf_kriva = dist.pdf(x, *params)
    plt.plot(x, pdf_kriva, 'r-', linewidth=3, label=f"Teorijska {oznaka} fittovana")

    # Oznaka aritmetičke sredine
    sredina = podaci.mean()
    plt.axvline(sredina, color='green', linestyle='--', linewidth=2, label=f"Srednja vrijednost uzorka ({sredina:.2f})")

    kratki_naziv = naziv_kolone.split('.')[-1].strip()[:40]
    plt.title(f"Teorijski fit i empirijska gustina za: {kratki_naziv}", fontsize=11, fontweight='bold', pad=12)
    plt.xlabel(naziv_kolone, fontsize=9)
    plt.ylabel("Gustoća (Density)", fontsize=9)
    plt.legend(fontsize='small', loc='upper left')
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()

    # 3. Chi-Square Goodness of Fit Test
    # Kreiranje 10 jednakih binova na osnovu opsega podataka
    kategorije = pd.cut(podaci, bins=10)
    observed_freq = kategorije.value_counts().sort_index().values

    bin_edges = np.histogram_bin_edges(podaci, bins=10)
    expected_freq = []

    # Računanje očekivanih frekvencija integracijom CDF-a kroz binove
    for i in range(len(bin_edges)-1):
        p_bin = dist.cdf(bin_edges[i+1], *params) - dist.cdf(bin_edges[i], *params)
        expected_freq.append(p_bin * len(podaci))

    expected_freq = np.array(expected_freq)

    # Izbjegavanje nula u očekivanim frekvencijama radi stabilnosti testa
    expected_freq = np.where(expected_freq == 0, 1e-5, expected_freq)

    # Normalizacija očekivanih frekvencija da suma bude jednaka opaženim
    expected_freq = expected_freq * (sum(observed_freq) / sum(expected_freq))

    # Podešavanje stepeni slobode (ddof) jer procjenjujemo parametre iz uzorka
    kriticni_ddof = Br_procijenjenih_param
    stepeni_slobode = (len(observed_freq) - 1) - kriticni_ddof

    # Pokretanje chi-square testa
    chi_stat, p_val = stats.chisquare(f_obs=observed_freq, f_exp=expected_freq, ddof=kriticni_ddof)

    print(f"\n" + "="*70)
    print(f" STATISTIČKI REZULTATI ZA VARIJABLU: {naziv_kolone}")
    print(f"="*70)
    print(f"Usvojeni nivo značajnosti (alfa): 0.05")
    print(f"Broj kategorija (binova): {len(observed_freq)}")
    print(f"Korigovani stepeni slobode (dof): {stepeni_slobode}")
    print(f"Nulta hipoteza (H0): Podaci prate teorijsku {oznaka} distribuciju.")
    print(f"Alternativna hipoteza (H1): Podaci NE prate teorijsku {oznaka} distribuciju.")
    print(f"-"*70)
    print(f"Izračunata Chi-Square statistika = {chi_stat:.4f}")
    print(f"Dobijena p-vrijednost = {p_val:.4e}")
    print(f"-"*70)

    if p_val < 0.05:
        print("ZAKLJUČAK: Odbacujemo nultu hipotezu (H0).")
        print(f"Podaci statistički značajno odstupaju od teorijske {oznaka} distribucije.")
    else:
        print("ZAKLJUČAK: Ne možemo odbaciti nultu hipotezu (H0).")
        print(f"Teorijska {oznaka} distribucija pruža adekvatan i dobar fit za ove podatke!")
    print("="*70 + "\n")

# Pokretanje analize za GPA (Normalna distribucija)
fituj_i_testiraj(df, kolona_gpa, "norm")

# Pokretanje analize za Sate učenja (Log-normalna distribucija)
fituj_i_testiraj(df, kolona_sati, "lognorm")
```

Nakon što smo pokrenule kod, procijenile parametre iz uzorka i izvršile formalni Chi-square Goodness of Fit test uz korekciju stepena slobode ($ddof=2$), suočile smo se sa ishodima koji su u potpunosti demantovali naša početna predviđanja.
### 1. Varijabla:
Moje predviđanje da će podaci pratiti Normalnu distribuciju i da će test to potvrditi je odbačeno od strane Chi-square testa, uz izuzetno nisku p-vrijednost ($p < 0.05$). ## 2. Testovi normalnosti-Fatima Fulurija

## PREDVIĐANJE(prije analize)

Za varijablu "Your current GPA (grade point average)" predviđam da će i Kolmogorov-Smirnov (KS) i Shapiro-Wilk testovi odbaciti nultu hipotezu o normalnosti ($p < 0.05$). Iako akademski prosjeci teoretski teže zvonastom obliku, stvarni podaci o ocjenama studenata najčešće imaju lijevu (negativnu) asimetriju jer dominiraju više ocjene, a postoji i strogo gornje ograničenje na 10.0. Očekujem da će QQ-plot prikazati odstupanja plavih tačaka od crvene linije na samim krajevima (repovima) distribucije.Za varijablu "How many hours per week do you spend on university-related activities..." sa sigurnošću predviđam da će oba formalna testa glatko odbaciti normalnost ($p < 0.05$). Broj sati učenja je tipična desno asimetrična varijabla sa izraženim "outlierima" (studenti koji uče ekstremno mnogo), što je nespojivo sa simetrijom normalne distribucije. Očekujem da će QQ-plot imati oblik krive koja se na desnom kraju naglo savija prema gore, bježeći od teorijske crvene linije.Što se tiče poređenja testova, očekujem da će Shapiro-Wilk test biti strožiji i dati još manju $p$-vrijednost od KS testa, jer Shapiro-Wilk posjeduje najveću statističku snagu za detekciju odstupanja od normalnosti na uzorcima ove veličine.

### Hipoteze za oba testa:
$H_0$ (Nulta hipoteza): Podaci prate normalnu distribuciju.

$H_1$ (Alternativna hipoteza): Podaci statistički značajno odstupaju od normalne distribucije.

Nivo značajnosti: $\alpha = 0.05$

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.stats.diagnostic import lilliefors  # Dodajemo zvanični Lilliefors test

# Nazivi kolona iz vašeg dataseta
kolona_gpa = "GPA"
kolona_sati = "Sati_ucenja"

def testiraj_normalnost(df_izvor, naziv_kolone, kratki_naziv_za_grafik):
    # Čišćenje podataka od NaN vrijednosti i pretvaranje u numerički tip
    podaci = pd.to_numeric(df_izvor[naziv_kolone], errors='coerce').dropna()

    print("\n" + "="*75)
    print(f" STATISTIČKI TESTOVI NORMALNOSTI ZA VARIJABLU: {kratki_naziv_za_grafik.upper()}")
    print("="*75)
    print(f"Nulta hipoteza (H0): Podaci prate normalnu distribuciju.")
    print(f"Alternativna hipoteza (H1): Podaci NE prate normalnu distribuciju.")
    print(f"Nivo značajnosti (alpha) = 0.05")
    print("-" * 75)

    # 1. Shapiro-Wilk Test
    shapiro_stat, shapiro_p = stats.shapiro(podaci)
    print(f"1. Shapiro-Wilk test (Izuzetno pouzdan za uzorke N < 2000):")
    print(f"   - Test statistika (W) = {shapiro_stat:.4f}")
    print(f"   - p-vrijednost = {shapiro_p:.4e}")
    if shapiro_p < 0.05:
        print("   -> ZAKLJUČAK: Odbacujemo H0 (Podaci NISU normalno distribuirani).")
    else:
        print("   -> ZAKLJUČAK: Ne možemo odbaciti H0 (Podaci prate normalnu distribuciju).")

    print("-" * 75)

    # 2. Kolmogorov-Smirnov Test sa Lilliefors korekcijom
    # (Pravilna modifikacija KS testa kada su parametri populacije nepoznati)
    ks_stat, ks_p = lilliefors(podaci, dist='norm')
    print(f"2. Kolmogorov-Smirnov test (sa Lilliefors korekcijom):")
    print(f"   - Test statistika (D) = {ks_stat:.4f}")
    print(f"   - p-vrijednost = {ks_p:.4e}")
    if ks_p < 0.05:
        print("   -> ZAKLJUČAK: Odbacujemo H0 (Podaci NISU normalno distribuirani).")
    else:
        print("   -> ZAKLJUČAK: Ne možemo odbaciti H0 (Podaci prate normalnu distribuciju).")

    print("="*75)

    # 3. Grafički prikaz: Unaprijeđeni QQ-Plot
    fig, ax = plt.subplots(figsize=(6, 4.5))
    sm.qqplot(podaci, line='s', ax=ax, color='#1f77b4', alpha=0.7)

    # Estetska dotjerivanja grafikona
    ax.set_title(f"QQ-Plot za varijablu: {kratki_naziv_za_grafik}", fontsize=12, fontweight='bold', pad=10)
    ax.set_xlabel("Teoretski kvantili (Normalna distribucija)", fontsize=10)
    ax.set_ylabel("Uzorčani (Empirijski) kvantili", fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
    print("\n")

# Pokretanje testova za GPA sa čistim nazivom za grafik
testiraj_normalnost(df, kolona_gpa, "Trenutni GPA (Prosjek ocjena)")

# Pokretanje testova za Sate učenja sa čistim nazivom za grafik
testiraj_normalnost(df, kolona_sati, "Sati učenja sedmično")
```

### Provjera (nakon analize)
Na generisanom QQ-plotu za varijablu Trenutni GPA, uočavamo da tačke prate crvenu referentnu liniju u centralnom dijelu, ali dolazi do devijacije na samim ekstremima.
Na gornjem desnom kraju tačke se primjetno zaravnjuju i padaju ispod linije, što vizuelno dokazuje efekt gornje granice (maksimalni prosjek 10.0 preko kojeg podaci fizički ne mogu ići). Na donjem lijevom kraju, tačke bježe iznad linije, što implicira brži pad frekvencija nego kod idealnog repa normalne distribucije, odnosno lijevu asimetriju. Zbog ovih geometrijskih odstupanja u repovima, formalni testovi su odbacili nultu hipotezu.
### Formalni statistički rezultati za Trenutni GPA
**Shapiro-Wilk test:** Test statistika $W = 0.9246$, p-vrijednost $= 2.35 \times 10^{-6}$ (odnosno $0.00000235$). Budući da je $p < 0.05$, nultu hipotezu odbacujemo i potvrđujemo da podaci statistički značajno odstupaju od normalne distribucije.

**Kolmogorov-Smirnov test (sa Lilliefors korekcijom):** Test statistika $D = 0.1216$, p-vrijednost $= 0.0010$. Pošto je $p < 0.05$, i ovaj test opravdano odbacuje nultu hipotezu.Konačni zaključak za GPA: Oba formalna testa su jednoglasno odbacila normalnost, što se savršeno poklapa sa našim početnim predviđanjem o negativnoj asimetriji i uticaju gornje granice ocjena na fakultetu.
### Analiza QQ-plota za Sate učenja sedmično

QQ-plot za Sate učenja sedmično pokazuje drastično strukturalno odstupanje koje podsjeća na oblik slova S ili zakrivljene parabole.Na desnoj strani grafikona (visoki kvantili), plave tačke rapidno eskaliraju visoko iznad crvene pravolinijske putanje. To je klasični vizuelni indikator teškog desnog repa (heavy-tailed distribution) i jake desne asimetrije. Podaci očigledno sadrže ekstremne "outliere" (studente koji uče preko 40 ili 50 sati), što u potpunosti uništava simetriju neophodnu za normalnu raspodjelu.
### Formalni statistički rezultati za Sate učenja sedmično
Shapiro-Wilk test: Test statistika $W = 0.9768$, p-vrijednost $= 0.0271$. S obzirom na to da je p-vrijednost manja od nivoa značajnosti ($\alpha = 0.05$), ovaj test uspješno odbacuje nultu hipotezu i zaključuje da podaci ne prate normalnu distribuciju.

Kolmogorov-Smirnov test (sa Lilliefors korekcijom): Test statistika $D = 0.0742$, p-vrijednost $= 0.1141$. Budući da je ovdje $p > 0.05$, ovaj test ne može odbaciti nultu hipotezu o normalnosti.
### Objašnjenje razlika u zaključcima testova
U slučaju varijable za sate učenja, našli smo se u specifičnoj situaciji gdje su testovi pružili različite zaključke. To se objašnjava fundamentalnim razlikama u njihovoj matematičkoj formulaciji i statističkoj snazi:

Statistička snaga (Power of the test): Shapiro-Wilk test je namjenski konstruisan isključivo za testiranje normalnosti i posjeduje daleko najveću statističku snagu. On direktno analizira korelaciju između podataka i idealnih normalnih kvantila. Sa druge strane, KS test je opšti test zasnovan na maksimalnoj vertikalnoj udaljenosti između empirijske i teorijske kumulativne distribucije ($D$ statistika). Zbog toga je KS test često "konzervativan" i "trom" na uzorcima srednje veličine, te lakše napravi grešku tipa II (ne odbaci netačnu $H_0$).

Osjetljivost na repove u odnosu na centar: KS test je najosjetljiviji oko medijana (sredine distribucije), dok je prilično "slijep" za odstupanja koja se dešavaju duboko u repovima. Suprotno tome, Shapiro-Wilk i QQ-plot ekstremno oštro reaguju na anomalije u repovima i prisustvo autlijera. Kod varijabli poput studentskih sati učenja (gdje je centralni dio relativno stabilan, ali repovi odstupaju zbog studenata sa ekstremno visokim brojem sati), Shapiro-Wilk test je ispravno prepoznao asimetriju i odbacio normalnost ($p = 0.0271$). Osnovni KS test sa Lilliefors korekcijom je podbacio zbog manje snage i pokazao lažnu tolerantnost ($p = 0.1141$). Ovo jasno dokazuje zašto je Shapiro-Wilk test superiorniji i pouzdaniji izbor za uzorke ove veličine.

# **3. Intervali pouzdanosti za srednje vrijednosti-Elmedina Tajić**

# PREDVIĐANJE (prije analize)

Za procjenu intervala pouzdanosti koristit ćemo t-statistiku za sve odabrane numeričke varijable. Iako naš uzorak sadrži $N = 128$ ispitanika (što je formalno veliki uzorak), varijanse populacija su nam potpuno nepoznate, što je standardna pojava u realnim anketnim istraživanjima. U takvim uslovima, matematički je opravdano koristiti kritične vrijednosti t-distribucije ($t$-statistiku) sa $n-1$ stepeni slobode, jer ona pruža adekvatnu korekciju i obuhvata dodatni varijabilitet zbog procjene standardne devijacije iz uzorka.

Za varijablu Trenutni GPA očekujemo relativno uzak interval pouzdanosti. Pretpostavljamo da su ocjene studenata prirodno homogene i grupisane oko centralnih akademskih vrijednosti, bez ekstremnih oscilacija na nivou cijelog uzorka.

Za varijablu Broj sati sedmično posvećenih fakultetskim aktivnostima očekujemo primjetno širi interval pouzdanosti. Količina vremena koju studenti ulažu varira u zavisnosti od njihove motivacije, težine smjera i lične organizacije, što stvara veću varijansu.

Za varijablu Broj učestvovanja u istraživanjima tokom protekle godine takođe očekujemo širi interval. Realno je pretpostaviti da imamo polarizaciju: veliki broj studenata koji nikada nisu učestvovali (vrijednost 0) i manji broj studenata koji su bili izrazito aktivni.


```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# TAČNI NAZIVI KOLONA DIREKTNO IZ TVOG DATASETA
pun_naziv_gpa = "4. Your current GPA (grade point average)"
pun_naziv_sati = "5. How many hours per week do you spend on university-related activities (lectures, studying, projects)?"
pun_naziv_anketa = "7. If yes, how many times in the past year?"

varijable = {
    pun_naziv_gpa: "Trenutni GPA (Prosjek ocjena)",
    pun_naziv_sati: "Sati učenja/aktivnosti sedmično",
    pun_naziv_anketa: "Broj učestvovanja u istraživanjima u zadnjoj godini"
}

rezultati = {}
for kolona, opis in varijable.items():
    podaci = pd.to_numeric(df[kolona], errors='coerce').dropna()
    n = len(podaci)
    sredina = np.mean(podaci)
    sem = stats.sem(podaci)
    ci95 = stats.t.interval(confidence=0.95, df=n-1, loc=sredina, scale=sem)
    rezultati[kolona] = {'n': n, 'mean': sredina, 'sem': sem, 'ci95': ci95}

# Posebni dio za GPA nivoe pouzdanosti
gpa_podaci = pd.to_numeric(df[pun_naziv_gpa], errors='coerce').dropna()
n_gpa = len(gpa_podaci)
mean_gpa = np.mean(gpa_podaci)
sem_gpa = stats.sem(gpa_podaci)
ci90 = stats.t.interval(0.90, n_gpa-1, loc=mean_gpa, scale=sem_gpa)
ci99 = stats.t.interval(0.99, n_gpa-1, loc=mean_gpa, scale=sem_gpa)

# Ispis rezultata u konzoli
print("="*70)
print("1. REZULTATI ZA 95% INTERVALE POUZDANOSTI")
print("="*70)
for kolona, opis in varijable.items():
    res = rezultati[kolona]
    print(f"\nVARIJABLA: {opis}")
    print(f"Veličina uzorka (n) = {res['n']}")
    print(f"Uzorčana srednja vrijednost (X_bar) = {res['mean']:.3f}")
    print(f"Standardna greška (SEM) = {res['sem']:.3f}")
    print(f"95% Interval pouzdanosti (CI) = [{res['ci95'][0]:.3f}, {res['ci95'][1]:.3f}]")
    print("-" * 50)

print("\n" + "====="*10)
print("2. POREĐENJE INTERVALA ZA VARIJABLU: GPA")
print("====="*10)
print(f"90% Interval pouzdanosti: [{ci90[0]:.3f}, {ci90[1]:.3f}]  -> Širina: {ci90[1]-ci90[0]:.3f}")
print(f"95% Interval pouzdanosti: [{rezultati[pun_naziv_gpa]['ci95'][0]:.3f}, {rezultati[pun_naziv_gpa]['ci95'][1]:.3f}]  -> Širina: {rezultati[pun_naziv_gpa]['ci95'][1]-rezultati[pun_naziv_gpa]['ci95'][0]:.3f}")
print(f"99% Interval pouzdanosti: [{ci99[0]:.3f}, {ci99[1]:.3f}]  -> Širina: {ci99[1]-ci99[0]:.3f}")
print("====="*10)

# Kreiranje subplota
fig, osi = plt.subplots(3, 1, figsize=(10, 10))
sns.set_theme(style="whitegrid")

# 1. Grafik: Sati učenja
sati_res = rezultati[pun_naziv_sati]
osi[0].errorbar(sati_res['mean'], 0, xerr=(sati_res['ci95'][1]-sati_res['ci95'][0])/2,
                fmt='o', color='#2196f3', markersize=10, linewidth=3, capsize=8)
osi[0].axvline(sati_res['mean'], color='gray', linestyle='--', alpha=0.5)
osi[0].set_yticks([])
osi[0].set_title("95% Interval pouzdanosti za: Sati učenja sedmično", fontsize=11, fontweight='bold')
osi[0].set_xlabel(f"Raspon sati (Srednja vrijednost = {sati_res['mean']:.2f})", fontsize=9)

# 2. Grafik: Broj učestvovanja u istraživanjima
ankete_res = rezultati[pun_naziv_anketa]
osi[1].errorbar(ankete_res['mean'], 0, xerr=(ankete_res['ci95'][1]-ankete_res['ci95'][0])/2,
                fmt='o', color='#4caf50', markersize=10, linewidth=3, capsize=8)
osi[1].axvline(ankete_res['mean'], color='gray', linestyle='--', alpha=0.5)
osi[1].set_yticks([])
osi[1].set_title("95% Interval pouzdanosti za: Broj učestvovanja u istraživanjima (Zadnja godina)", fontsize=11, fontweight='bold')
osi[1].set_xlabel(f"Raspon broja učešća (Srednja vrijednost = {ankete_res['mean']:.2f})", fontsize=9)

# 3. Grafik: GPA poređenje širine intervala
err_90 = (ci90[1] - ci90[0]) / 2
err_95 = (rezultati[pun_naziv_gpa]['ci95'][1] - rezultati[pun_naziv_gpa]['ci95'][0]) / 2
err_99 = (ci99[1] - ci99[0]) / 2

nivoi = ['90% Pouzdanost', '95% Pouzdanost', '99% Pouzdanost']
sredine_gpa = [mean_gpa, mean_gpa, mean_gpa]
greske_gpa = [err_90, err_95, err_99]
boje_gpa = ['#ff9800', '#2196f3', '#f44336']

for i in range(3):
    osi[2].errorbar(sredine_gpa[i], i, xerr=greske_gpa[i], fmt='o', color=boje_gpa[i],
                    markersize=9, linewidth=3, capsize=8)

osi[2].axvline(mean_gpa, color='gray', linestyle='--', alpha=0.5)
osi[2].set_yticks([0, 1, 2])
osi[2].set_yticklabels(nivoi, fontsize=9)
osi[2].set_ylim(-0.5, 2.5)
osi[2].set_title("Poređenje širine intervala zavisno od nivoa pouzdanosti (GPA)", fontsize=11, fontweight='bold')
osi[2].set_xlabel(f"Raspon GPA (Srednja vrijednost = {mean_gpa:.2f})", fontsize=9)

plt.tight_layout(pad=3.0)
plt.show()
```

# **4. Intervali pouzdanosti za proporcije-Elmedina Tajić**



# PREDVIĐANJE (prije analize)

Za analizu proporcija u našem istraživanju odabrale smo dvije ključne kategorijske varijable koje direktno ispituju granice komfora i spremnost studenata na specifične, tehnološki invazivnije oblike prikupljanja podataka: Screen recording (snimanje ekrana tokom korištenja aplikacije) i Eye-tracking (biometrijsko praćenje pokreta očiju).

S obzirom na to da radimo sa uzorkom od $N = 128$ ispitanika, za procjenu intervala pouzdanosti svake kategorije koristit ćemo Wilsonovu metodu (Wilson score interval). Ova statistička metoda je izabrana jer je dokazano robusnija i preciznija od standardne Waldove metode, posebno kada analiziramo proporcije unutar manjih podkategorija ili kada su odgovori asimetrični, čime osiguravamo da donje granice intervala ne odu u matematički nemoguće negativne vrijednosti.

Za varijablu koja se odnosi na snimanje ekrana (Screen recording), predviđamo da će uvjerljivo najveći broj studenata izabrati neutralnu ili uslovnu opciju, odnosno odgovor "Možda, zavisi od konteksta".Naša pretpostavka je da većina studenata iz tehničke populacije nema apriori odbojnost prema samom istraživačkom procesu, ali zbog opravdane svijesti o digitalnoj privatnosti i sigurnosti podataka ne žele dati blanko pristanak. Oni žele precizne informacije o tome koja se aplikacija snima i u koje svrhe. Procjenjujemo da će uzorčana proporcija ove kategorije biti oko $p = 0.50$ ($50\%$). Budući da se radi o umjerenom uzorku i proporciji koja maksimizira varijansu, očekujemo stabilan, srednje širok 95% interval pouzdanosti koji će se kretati otprilike u rasponu od $[0.40, 0.60]$.

Za varijablu koja ispituje spremnost na praćenje pokreta očiju (Eye-tracking), predviđamo da će studenti pokazati znatno veći nivo opreza, skepse i distanciranosti.Eye-tracking je hardverski i biometrijski specifičnija metoda koja zahtijeva viši nivo direktnog, fizičkog angažmana i posmatranja ispitanika. Zbog toga očekujemo da će proporcija čistih, eksplicitnih odgovora "Da" biti primjetno niža u odnosu na uobičajene oblike anketa i da će iznositi oko $p = 0.30$ ($30\%$). Budući da ova proporcija bježi od centralnih $0.50$ prema nuli, matematički očekujemo nešto užu i precizniju marginu greške, sa procijenjenim 95% intervalom pouzdanosti u okvirnom rasponu od $[0.22, 0.38]$.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.proportion import proportion_confint

# 1. Definišemo tačne kolone iz tvog preimenovanog df-a koje analiziramo
# (Prilagodi nazive ako se u tvom df-u zovu nijansu drugačije)
odabrane_kolone = {
    "Spremnost_duze_ankete": "Spremnost za Screen Recording / Duže ankete",
    "Neutralnost_u_anketama": "Spremnost za Eye-Tracking / Specifične metode"
}

sns.set_theme(style="whitegrid")

# Prolazimo kroz svaku varijablu
for kolona, puni_naziv in odabrane_kolone.items():
    if kolona not in df.columns:
        print(f"Upozorenje: Kolona '{kolona}' nije pronađena u df-u. Provjerite naziv.")
        continue

    print("\n" + "="*85)
    print(f" ANALIZA INTERVALA POUZDANOSTI ZA PROPORCIJE: {puni_naziv.upper()}")
    print("="*85)

    counts = df[kolona].value_counts().dropna()
    n = counts.sum()

    print(f"Ukupan broj važećih odgovora (N) = {n}")
    print(f"Nivo značajnosti (alpha) = 0.05 | Metoda: Wilson Score Interval")
    print("-" * 85)

    # Liste u koje ćemo spasiti podatke za crtanje grafikona
    kategorije_lista = []
    proporcije_lista = []
    greske_donja = []
    greske_gornja = []

    # Računanje intervala pouzdanosti za SVAKU kategoriju unutar varijable
    for kategorija, broj in counts.items():
        p = broj / n

        # Wilsonov interval je zlatni standard za proporcije
        ci_low, ci_high = proportion_confint(
            count=broj,
            nobs=n,
            alpha=0.05,
            method='wilson'
        )

        # Spremamo podatke za vizuelizaciju
        kategorije_lista.append(kategorija)
        proporcije_lista.append(p)
        greske_donja.append(p - ci_low)
        greske_gornja.append(ci_high - p)

        print(f"Kategorija: '{kategorija}'")
        print(f"   - Frekvencija = {broj} od {n}")
        print(f"   - Uzorčana proporcija (p) = {p:.4f} ({p*100:.2f}%)")
        print(f"   - 95% Interval pouzdanosti = [{ci_low:.4f}, {ci_high:.4f}]")
        print(f"   - Raspon (Širina intervala) = {ci_high - ci_low:.4f}")
        print("-" * 60)

    # 2. GRAFIČKI PRIKAZ: Izrada Error-Bar grafikona za intervale pouzdanosti
    plt.figure(figsize=(8, 4))

    # Asimetrične greške za errorbar (donje i gornje odstupanje od tačke p)
    asimetricna_greska = np.array([greske_donja, greske_gornja])

    # Crta tačke (proporcije) i linije (interval pouzdanosti)
    plt.errorbar(
        x=proporcije_lista,
        y=kategorije_lista,
        xerr=asimetricna_greska,
        fmt='o',
        color='#2E75B6',
        ecolor='#A6A6A6',
        elinewidth=3,
        capsize=6,
        markersize=8,
        label='95% Interval pouzdanosti'
    )

    # Dodavanje vertikalne linije na 0.50 radi lakše analize teorijskog maksimuma varijanse
    plt.axvline(x=0.5, color='red', linestyle=':', alpha=0.5, label='Teorijski maksimum (p=0.5)')

    plt.title(f"95% CI za proporcije: {kolona}\n({puni_naziv})", fontsize=11, fontweight='bold', pad=10)
    plt.xlabel("Proporcija (Udio u uzorku)", fontsize=10)
    plt.ylabel("Kategorije odgovora", fontsize=10)
    plt.xlim(-0.05, 1.05) # Raspon od 0 do 100% sa malo lufta
    plt.legend(loc="lower right")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()
    print("\n" + "="*85)
```

# **5. Chi-square test nezavisnosti - Emina Fejzagić**

## **PREDVIĐANJE (prije analize)**


Za ovaj zadatak izabrale smo tri para kategorijskih varijabli kako bismo ispitale njihovu međusobnu zavisnost pomoću Chi-square testa nezavisnosti.

1. Par: Spol ("1. Gender") i Pristanak na snimanje ekrana ("21. Would you agree to participate in research that involves screen recording while you use an application?")

Očekivanje: Intuitivno pretpostavljamo da bi ove dvije varijable mogle biti zavisne. Pitanja privatnosti na internetu često izazivaju različite reakcije, a moguće je da su žene opreznije kada je u pitanju ustupanje pristupa i snimanje njihovog ličnog uređaja u odnosu na muškarce.

Predviđanje p-vrijednosti: Očekujemo statistički značajnu zavisnost, odnosno da će p-vrijednost biti manja od nivoa značajnosti (p < 0.05).

2. Par: Rad uz studij ("3. Are you currently working alongside your studies?") i Prethodno učešće u istraživanjima ("6. Have you ever participated in any research or study as a respondent...")

Očekivanje: Mislimo da su ove dvije varijable potpuno nezavisne. To da li student radi ili ne vjerovatno nema nikakvog direktnog uticaja na to da li ga je neko nekada zamolio da ispuni anketu (što se najčešće dešava na fakultetu ili preko društvenih mreža, neovisno o poslu).

Predviđanje p-vrijednosti: Očekujemo da nećemo odbaciti nultu hipotezu i da neće biti statistički značajne zavisnosti (p > 0.05).

3. Par: Spol ("1. Gender") i Rad uz studij ("3. Are you currently working alongside your studies?")

Očekivanje: Očekujemo da su i ove varijable nezavisne. Na današnjem tržištu rada i muškim i ženskim studentima su podjednako potrebni (i dostupni) studentski poslovi. Ne vidimo razlog zašto bi jedan spol statistički značajno više radio uz fakultet.

Predviđanje p-vrijednosti: Očekujemo da nema zavisnosti (p > 0.05).


```python
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje dataseta
df = pd.read_csv('Zadaca 1 - Dataset (1).csv', encoding='latin1')

# Prilagodi ove nazive kolona ako si ih u svom glavnom df-u već skratila!
kol_spol = '1. Gender'
kol_snimanje = '21. Would you agree to participate in research that involves screen recording while you use an application?'
kol_rad = '3. Are you currently working alongside your studies?'
kol_ucesce = '6. Have you ever participated in any research or study as a respondent (survey, interview, app testing, experiment)?'

parovi = [
    (kol_spol, kol_snimanje, "Par 1: Spol i Snimanje ekrana"),
    (kol_rad, kol_ucesce, "Par 2: Rad uz studij i Prethodno učešće"),
    (kol_spol, kol_rad, "Par 3: Spol i Rad uz studij")
]

alpha = 0.05  # Nivo značajnosti
sns.set_theme(style="whitegrid")

for var1, var2, naziv in parovi:
    print("\n" + "="*80)
    print(f" CHI-SQUARE TEST NEZAVISNOSTI: {naziv.upper()}")
    print("="*80 + "\n")

    # 1. Kreiranje kontingencijske tabele
    contingency_table = pd.crosstab(df[var1], df[var2])
    print("1. Kontingencijska tabela (Opažene frekvencije):")
    print(contingency_table)
    print("-" * 50)

    # 2. Izvođenje Chi-square testa
    chi2_stat, p_val, dof, expected = stats.chi2_contingency(contingency_table)

    print("2. Hipoteze:")
    print("   H0: Varijable su statistički nezavisne (nema povezanosti u populaciji).")
    print("   H1: Varijable su statistički zavisne (postoji povezanost u populaciji).")
    print(f"   Nivo značajnosti (alpha) = {alpha}\n")

    print("3. Rezultati testa:")
    print(f"   - Test statistika (Chi-square) = {chi2_stat:.4f}")
    print(f"   - Stepeni slobode (dof) = {dof}")
    print(f"   - p-vrijednost = {p_val:.4e}")
    print("-" * 50)

    # 3. Provjera pretpostavki testa (Cochranovo pravilo)
    print("4. Provjera pretpostavki (Očekivane frekvencije):")
    nizak_broj = (expected < 5).sum()
    ukupno_celija = expected.size
    procenat_niskih = (nizak_broj / ukupno_celija) * 100

    print(f"   - Broj ćelija sa očekivanom frekvencijom < 5: {nizak_broj} od {ukupno_celija} ({procenat_niskih:.1f}%)")
    if procenat_niskih <= 20:
        print("   -> PRETPOSTAVKA ZADOVOLJENA: Manje od 20% ćelija ima očekivanu vrijednost < 5.")
    else:
        print("   -> UPOZORENJE: Preko 20% ćelija ima očekivanu vrijednost < 5. Pouzdanost može biti ugrožena!")
    print("-" * 50)

    # 4. Zaključak
    print("5. Statistički zaključak:")
    if p_val < alpha:
        print(f"   Pošto je p-vrijednost ({p_val:.4f}) < {alpha} -> ODBACUJEMO H0.")
        print("   Zaključak: Postoji statistički značajna zavisnost između posmatranih varijabili.")
    else:
        print(f"   Pošto je p-vrijednost ({p_val:.4f}) >= {alpha} -> NE MOŽEMO odbaciti H0.")
        print("   Zaključak: Nemamo dovoljno dokaza o povezanosti; varijable smatramo nezavisnim.")
    print("-" * 50)

    # 5. VIZUELIZACIJA: Grupisani Bar Chart za vizuelnu provjeru zavisnosti
    plt.figure(figsize=(8, 4.5))
    df_plot = df[[var1, var2]].dropna()

    sns.countplot(x=var1, hue=var2, data=df_plot, palette="muted", edgecolor="0.2")

    # Prilagođavanje prikaza grafikona
    kratki_naziv1 = var1.split('.')[-1].strip()[:25]
    kratki_naziv2 = var2.split('.')[-1].strip()[:25]
    plt.title(f"Distribucija odgovora: {kratki_naziv1} u funkciji {kratki_naziv2}", fontsize=11, fontweight='bold', pad=10)
    plt.xlabel(var1, fontsize=9)
    plt.ylabel("Broj studenata (Frekvencija)", fontsize=9)
    plt.legend(title=kratki_naziv2, loc="upper right", fontsize='small')
    plt.tight_layout()
    plt.show()
    print("\n" + "="*80 + "\n")
```

# **6. Testiranje hipoteza - Emina Fejzagić**

## **PREDVIĐANJE (prije analize)**



Za ovaj zadatak odlučile smo postaviti 4 različite hipoteze koje sagledavaju različite aspekte studentskog života i odnosa prema anketama. Pošto smo u Zadaći 1 otkrile (kroz testove normalnosti) da mnoge naše kontinuirane varijable ne prate strogo normalnu distribuciju, pažljivo smo birale testove, oslanjajući se više na robusne neparametarske metode.

**1. Hipoteza: Razlika u vremenu provedenom u učenju između studenata koji rade i ne rade**
* H₀: Srednji rangovi broja sati učenja ne razlikuju se između studenata koji rade i onih koji ne rade.
* H₁: Srednji rangovi broja sati učenja se razlikuju između ove dvije grupe.
* Zašto nas zanima: Želimo ispitati da li posao uz studij drastično "oduzima" vrijeme predviđeno za akademske obaveze.
* Test: **Mann-Whitney U test** . Biramo ga umjesto t-testa za nezavisne uzorke jer broj sati učenja najčešće nije normalno raspoređen.
* Predviđanje: Očekujemo da će H₀ biti odbačena. Mislimo da studenti koji rade imaju znatno manje vremena za učenje.

**2. Hipoteza: Povezanost (korelacija) između truda i ocjena**
* H₀: Ne postoji povezanost između broja sati provedenih u učenju i prosječne ocjene (GPA).
* H₁: Postoji pozitivna povezanost između broja sati i prosječne ocjene.
* Zašto nas zanima: Želimo provjeriti staro studentsko pravilo "više truda = bolje ocjene" na našem uzorku.
* Test: **Spearmanov test korelacije** (jednosmjerni). Pravi izbor jer radimo sa varijablama koje vjerovatno odstupaju od normalnosti.
* Predviđanje: Očekujemo da će H₀ biti odbačena i da ćemo dokazati pozitivnu korelaciju.

**3. Hipoteza: Strpljenje za ankete u zavisnosti od godine studija**
* H₀: Nema razlike u maksimalnom broju minuta koje su studenti spremni izdvojiti za anketu bez naknade s obzirom na to koja su godina studija.
* H₁: Barem jedna godina studija se statistički značajno razlikuje od ostalih po vremenu koje bi izdvojili.
* Zašto nas zanima: Zanima nas da li se "empatija" prema istraživačima mijenja kako studenti stare.
* Test: **Kruskal-Wallis test** (neparametarska alternativa ANOVA-i). Koristimo ga jer upoređujemo više od 2 grupe, a varijabla minuta nije normalno raspoređena.
* Predviđanje: Očekujemo da nećemo odbaciti H₀. Mislimo da strpljenje zavisi više od karaktera osobe.

**4. Hipoteza: Iznadprosječnost našeg uzorka (Da li ispitujemo najbolje studente?)**
* H₀: Prosječan GPA studenata u populaciji (našem uzorku) je jednak 8.0.
* H₁: Prosječan GPA u populaciji (našem uzorku) je veći od 8.0.
* Zašto nas zanima: Često se dešava "pristrasnost uzorka" gdje ankete popunjavaju odgovorniji studenti sa boljim ocjenama.
* Test: **T-test za jedan uzorak (Jednosmjerni)**.
* Predviđanje: Očekujemo da će H₀ biti odbačena. Naša intuicija je da prosječna ocjena premašuje 8.0.



```python
import pandas as pd
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje podataka uz ispravan encoding
df = pd.read_csv('Zadaca 1 - Dataset (1).csv', encoding='latin1')

# Definisanje kolona (Prilagodi nazive ako su u tvom df-u skraćeni!)
col_rad = '3. Are you currently working alongside your studies?'
col_sati = '5. How many hours per week do you spend on university-related activities (lectures, studying, projects)?'
col_gpa = '4. Your current GPA (grade point average)'
col_godina = '2. Year of study'
col_minute = '18. What is the maximum number of minutes you would be willing to spend on a single survey/study without compensation?'

# ČIŠĆENJE PODATAKA: Konverzija tekstualnih brojeva u prave brojeve, pretvaranje grešaka u NaN
df[col_sati] = pd.to_numeric(df[col_sati], errors='coerce')
df[col_gpa] = pd.to_numeric(df[col_gpa], errors='coerce')
df[col_minute] = pd.to_numeric(df[col_minute], errors='coerce')

# Uklanjanje redova koji imaju prazna polja za ove ključne kolone
df_clean = df.dropna(subset=[col_rad, col_sati, col_gpa, col_godina, col_minute]).copy()
alpha = 0.05

# Postavljanje platna za grafikone (2x2 raspored za 4 testa)
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
sns.set_theme(style="whitegrid")

print("===================================================================")
print("TEST 1: Mann-Whitney U test (Sati učenja vs. Rad uz studij)")
print("===================================================================")
sati_rade = df_clean[df_clean[col_rad] == 'Da'][col_sati]
sati_ne_rade = df_clean[df_clean[col_rad] == 'Ne'][col_sati]

stat_mwu, p_mwu = stats.mannwhitneyu(sati_rade, sati_ne_rade, alternative='two-sided')
print(f"Pretpostavke: Nezavisni uzorci, podaci nisu strogo normalno raspoređeni (korišteni rangovi).")
print(f"Test statistika (U): {stat_mwu:.4f}")
print(f"p-vrijednost: {p_mwu:.4f}")
if p_mwu < alpha:
    print("Zaključak: Odbacujemo H0. Postoji statistički značajna razlika u satima učenja između zaposlenih i nezaposlenih studenata.")
else:
    print("Zaključak: Ne odbacujemo H0. Nema dovoljno dokaza da rad utiče na broj sati učenja.")

# Grafikon 1: Boxplot za Mann-Whitney
sns.boxplot(ax=axes[0, 0], x=col_rad, y=col_sati, data=df_clean, palette="Set2")
axes[0, 0].set_title("Test 1: Sati učenja u odnosu na radni status", fontweight='bold', pad=10)
axes[0, 0].set_xlabel("Radi uz studij")
axes[0, 0].set_ylabel("Sati sedmično")


print("\n===================================================================")
print("TEST 2: Spearmanova korelacija (Sati učenja i GPA)")
print("===================================================================")
stat_corr, p_corr_two_sided = stats.spearmanr(df_clean[col_sati], df_clean[col_gpa])
p_corr = p_corr_two_sided / 2 if stat_corr > 0 else 1.0

print(f"Pretpostavke: Obje varijable su mjerene barem na ordinalnoj skali, odnos je monoton.")
print(f"Test statistika (Spearman Rho): {stat_corr:.4f}")
print(f"p-vrijednost (jednosmjerna): {p_corr:.4f}")
if p_corr < alpha:
    print("Zaključak: Odbacujemo H0. Postoji statistički značajna pozitivna korelacija između sati učenja i GPA.")
else:
    print("Zaključak: Ne odbacujemo H0. Sati učenja i GPA nisu statistički značajno pozitivno povezani.")

print(f"\n*Demonstracija različitog nivoa značajnosti (alpha = 0.01)*")
if p_corr < 0.01:
    print(f" Korelacija bi I DALJE bila značajna (p = {p_corr:.4f} < 0.01). Rezultat je izuzetno robustan.")
else:
    print(f" Odluka bi se PROMIJENILA! Pri p = {p_corr:.4f}, ne bismo mogle odbaciti H0 na nivou 0.01.")

# Grafikon 2: Regplot (Scatter + linija trenda) za Korelaciju
sns.regplot(ax=axes[0, 1], x=col_sati, y=col_gpa, data=df_clean, scatter_kws={'alpha':0.6, 'color':'#1f77b4'}, line_kws={'color':'red'})
axes[0, 1].set_title("Test 2: Korelacija između sati učenja i GPA", fontweight='bold', pad=10)
axes[0, 1].set_xlabel("Sati učenja sedmično")
axes[0, 1].set_ylabel("Trenutni GPA")


print("\n===================================================================")
print("TEST 3: Kruskal-Wallis test (Minute za anketu vs. Godina studija)")
print("===================================================================")
jedinstvene_godine = sorted(df_clean[col_godina].unique())
grupe_godina = [df_clean[df_clean[col_godina] == g][col_minute].values for g in jedinstvene_godine]
stat_kw, p_kw = stats.kruskal(*grupe_godina)

print(f"Pretpostavke: Nezavisni uzorci, kategorijska sa 3+ nivoa, podaci ne prate normalnu distribuciju.")
print(f"Test statistika (H): {stat_kw:.4f}")
print(f"p-vrijednost: {p_kw:.4f}")
if p_kw < alpha:
    print("Zaključak: Odbacujemo H0. Godina studija statistički značajno utiče na spremnost izdvajanja vremena za ankete.")
else:
    print("Zaključak: Ne odbacujemo H0. Nema statistički značajne razlike u strpljenju s obzirom na godinu studija.")

# Grafikon 3: Violinplot za Kruskal-Wallis
sns.violinplot(ax=axes[1, 0], x=col_godina, y=col_minute, data=df_clean, palette="Pastel1", order=jedinstvene_godine)
axes[1, 0].set_title("Test 3: Maksimalno minuta za anketu po godinama studija", fontweight='bold', pad=10)
axes[1, 0].set_xlabel("Godina studija")
axes[1, 0].set_ylabel("Maksimalno minuta")


print("\n===================================================================")
print("TEST 4: T-test za jedan uzorak (Da li je GPA veći od 8.0?)")
print("===================================================================")
stat_t, p_t_two_sided = stats.ttest_1samp(df_clean[col_gpa], popmean=8.0)
p_t = p_t_two_sided / 2 if stat_t > 0 else 1.0

print(f"Pretpostavke: Kontinuirani podaci, uzorak dovoljno veliki da teži normalnosti aritmetičke sredine (CLT).")
print(f"Test statistika (T): {stat_t:.4f}")
print(f"p-vrijednost (jednosmjerna): {p_t:.4f}")
if p_t < alpha:
    print("Zaključak: Odbacujemo H0. Prosjek (GPA) studenata u našem uzorku je statistički značajno veći od 8.0.")
else:
    print("Zaključak: Ne odbacujemo H0. Nemamo statističkih dokaza da je prosjek veći od 8.0.")

# Grafikon 4: Histogram distribucije GPA sa linijom istorijske/testirane srednje vrijednosti (8.0)
sns.histplot(ax=axes[1, 1], x=col_gpa, data=df_clean, kde=True, color="purple", alpha=0.5)
axes[1, 1].axvline(x=8.0, color='red', linestyle='--', linewidth=2.5, label='Testirana vrijednost (8.0)')
axes[1, 1].axvline(x=df_clean[col_gpa].mean(), color='green', linestyle='-', linewidth=2.5, label=f'Uzorčana sredina ({df_clean[col_gpa].mean():.2f})')
axes[1, 1].set_title("Test 4: Distribucija GPA u odnosu na konstantu 8.0", fontweight='bold', pad=10)
axes[1, 1].set_xlabel("GPA")
axes[1, 1].set_ylabel("Frekvencija")
axes[1, 1].legend()

# Podešavanje razmaka između grafikona da se naslovi i ose ne sudaraju
plt.tight_layout()
plt.subplots_adjust(wspace=0.35, hspace=0.45)

plt.show()
print("\n===================================================================")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 1. Definisani tačni nazivi kolona direktno iz tvog učitanog fajla
pun_naziv_gpa = "4. Your current GPA (grade point average)"
pun_naziv_sati = "5. How many hours per week do you spend on university-related activities (lectures, studying, projects)?"
pun_naziv_rad = "3. Are you currently working alongside your studies?"

# 2. Čišćenje i priprema podataka
df_clean = df.copy()
df_clean['GPA'] = pd.to_numeric(df_clean[pun_naziv_gpa], errors='coerce')
df_clean['Sati'] = pd.to_numeric(df_clean[pun_naziv_sati], errors='coerce')
df_clean = df_clean.dropna(subset=['GPA', 'Sati'])

# Prilagođavanje naziva kategorija za akademski prikaz
df_clean[pun_naziv_rad] = df_clean[pun_naziv_rad].replace({'Yes': 'Zaposleni studenti', 'No': 'Nezaposleni studenti'})

# Postavljanje profesionalnog stila (whitegrid je odličan za preglede)
sns.set_theme(style="whitegrid")

# Kreiranje figure sa 3 subplota u jednom redu
fig, osi = plt.subplots(1, 3, figsize=(18, 5.5))

# Dodavanje glavnog, formalnog naslova iznad svih grafikona
fig.suptitle("Grafički prikaz ključnih zaključaka analize", fontsize=14, fontweight='bold', y=0.98)

# =========================================================================
# 1. GRAFIK (Lijevo): Sati učenja u zavisnosti od radnog statusa
# =========================================================================
# Koristimo 'hue' parametar da izbjegnemo Seaborn upozorenja i dobijemo čiste boje
sns.boxplot(data=df_clean, x=pun_naziv_rad, y='Sati', ax=osi[0], hue=pun_naziv_rad, palette="muted", width=0.4, legend=False)
sns.stripplot(data=df_clean, x=pun_naziv_rad, y='Sati', ax=osi[0], color="black", alpha=0.25, jitter=0.08)
osi[0].set_title("Distribucija sati aktivnosti prema\nradnom statusu studenata", fontsize=11, fontweight='bold', pad=12)
osi[0].set_xlabel("Radni status studenta", fontsize=10)
osi[0].set_ylabel("Sati aktivnosti sedmično", fontsize=10)

# =========================================================================
# 2. GRAFIK (Sredina): Normalna raspodjela za GPA (Čisti akademski prikaz)
# =========================================================================
# Empirijski podaci (histogram)
sns.histplot(df_clean['GPA'], kde=False, stat="density", ax=osi[1], color="#4a90e2", alpha=0.6, bins=12, label="Empirijski podaci (Uzorak)")

# Teorijska normalna kriva
mu, std = df_clean['GPA'].mean(), df_clean['GPA'].std()
x = np.linspace(df_clean['GPA'].min(), df_clean['GPA'].max(), 100)
p = stats.norm.pdf(x, mu, std)
osi[1].plot(x, p, color="#d0021b", linestyle='--', linewidth=2, label="Teorijska normalna raspodjela")

osi[1].set_title("Poređenje empirijske distribucije GPA\ni teorijske normalne krive", fontsize=11, fontweight='bold', pad=12)
osi[1].set_xlabel("Trenutni GPA (Prosjek ocjena)", fontsize=10)
osi[1].set_ylabel("Gustoća vjerovatnoće", fontsize=10)
osi[1].legend(loc="upper left", fontsize=9)

# =========================================================================
# 3. GRAFIK (Desno): Ograničenja - Vizuelni prikaz zaokruživanja odgovora
# =========================================================================
# Analiza 6 najčešćih pojedinačnih odgovora
top_sati = df_clean['Sati'].value_counts().nlargest(6).sort_index()

# Pretvaramo indekse u stringove ("10 sati", "20 sati"...) da grafik bude simetričan i jasan
x_labels = [f"{int(iznos)} h" for iznos in top_sati.index]

# Crtanje ujednačenog, elegantnog barplota
sns.barplot(x=x_labels, y=top_sati.values, ax=osi[2], color="#4a5568")
osi[2].set_title("Analiza učestalosti najčešćih pojedinačnih\nodgovora za sedmične sati aktivnosti", fontsize=11, fontweight='bold', pad=12)
osi[2].set_xlabel("Prijavljene vrijednosti u anketama", fontsize=10)
osi[2].set_ylabel("Frekvencija (Broj studenata)", fontsize=10)

# Fino podešavanje razmaka da se naslovi i ose nigdje ne sudaraju
plt.tight_layout(rect=[0, 0, 1, 0.93])
plt.show()
```

## BONUS: Linearna regresija-Fatima Fulurija

### PREDVIĐANJE (Prije analize)

Sati_ucenja je zavisna varijabla koja je odabrana (Koliko sati sedmično student provodi na fakultetske aktivnosti). Izabrali smo je jer nas zanima šta najviše utiče na akademski angažman studenata.

GPA (Trenutni prosjek) je nezavisna varijabla koja je odabrana. Očekujemo da bude značajan i pozitivan prediktor, jer studenti koji više uče uglavnom ostvaruju viši prosjek ocjena.

Zaposlen (Da li student radi uz studij-druga nezavisna varijabla) Očekujemo da bude značajan i negativan prediktor, jer zaposleni studenti imaju manje slobodnog vremena na raspolaganju za učenje.

Očekivani $R^2$: Očekujemo nizak do umjeren $R^2$ (oko 0.15 – 0.25). Iako su prosjek i posao važni faktori, sati učenja zavise i od lične motivacije, težine samog fakulteta, organizacijskih vještina i drugih neizmjerenih faktora.



```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# 1. Učitavanje podataka
try:
    df = pd.read_csv('Zadaca 1 - Dataset (1).csv', encoding='windows-1250')
except UnicodeDecodeError:
    df = pd.read_csv('Zadaca 1 - Dataset (1).csv', encoding='latin1')

# 2. Skraćivanje kolona
df.columns = [
    'Timestamp', 'Pol', 'Godina_studija', 'Zaposlen', 'GPA', 'Sati_ucenja',
    'Ucestvovao_istrazivanju', 'Broj_ucestvovanja_godina', 'Tipovi_istrazivanja',
    'Iskrenost_odgovora', 'Obaveza_popunjavanja', 'Dijeljenje_drugima',
    'Cijenim_istrazivace', 'Zahvalnost_energiji', 'Nasumicno_popunjavanje',
    'Gubljenje_vremena', 'Ljutnja_ankete', 'Odbijanje_ankete',
    'Maks_minuta_bez_kompenzacije', 'Rank_Online', 'Rank_Laboratorijski',
    'Rank_Fokus_grupe', 'Rank_Intervju', 'Rank_Terensko', 'Motivacija',
    'Kompenzacija_uticaj', 'Spremnost_duze_ankete', 'Neutralnost_u_anketama',
    'Uticaj_dizajna', 'Zabrinutost_privatnost'
]

# 3. Čišćenje podataka
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce')
df['Sati_ucenja'] = pd.to_numeric(df['Sati_ucenja'], errors='coerce')
df['Zaposlen_Dummy'] = df['Zaposlen'].apply(lambda x: 1 if pd.notnull(x) and 'Da' in str(x) else 0)
df_clean = df[['Sati_ucenja', 'GPA', 'Zaposlen_Dummy']].dropna()

# 4. Pokretanje regresije
X = df_clean[['GPA', 'Zaposlen_Dummy']]
X = sm.add_constant(X)
y = df_clean['Sati_ucenja']
model = sm.OLS(y, X).fit()

# =========================================================================
# OVDJE JE SE KOD SPAJA SA PROVJEROM PRETPOSTAVKI I DIZAJNIRANOM TABELOM
# =========================================================================

# PRVI DIO: Grafikoni za pretpostavke (Linearnost i Normalnost)
reziduali = model.resid
sns.set_theme(style="whitegrid")
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=model.predict(), y=reziduali, color='purple', alpha=0.7)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Provjera linearnosti\n(Predviđene vrijednosti vs Reziduali)')
plt.xlabel('Predviđene vrijednosti (Sati učenja)')
plt.ylabel('Reziduali')

plt.subplot(1, 2, 2)
stats.probplot(reziduali, dist="norm", plot=plt)
plt.title('Provjera normalnosti\n(Q-Q Plot reziduala)')
plt.xlabel('Teoretski kvantili')
plt.ylabel('Prebrojani kvantili')

plt.tight_layout()
plt.show()

# DRUGI DIO: Fensi tabela koeficijenata
def znacajnost(p):
    if p < 0.01:   return 'Visoko značajna (p < 0.01)'
    elif p < 0.05: return 'Značajna (p < 0.05)'
    elif p < 0.10: return 'Djelimično značajna (na 10%)'
    else:          return 'Nije statistički značajna'

varijable = {
    'const':           'Konstanta (Intercept)',
    'GPA':             'Trenutni prosjek ocjena (GPA)',
    'Zaposlen_Dummy':  'Radni status (Zaposlen/Ne)'
}

izvjestaj_podaci = {
    'Varijabla u modelu':     [varijable[k] for k in varijable],
    'Koeficijent (β)':         [round(model.params[k], 4) for k in varijable],
    'Standardna greška':      [round(model.bse[k], 3)    for k in varijable],
    't-statistika':           [round(model.tvalues[k], 3) for k in varijable],
    'p-vrijednost (P>|t|)':   [round(model.pvalues[k], 3) for k in varijable],
    'Statistička značajnost': [znacajnost(model.pvalues[k]) for k in varijable],
}

tabela = pd.DataFrame(izvjestaj_podaci)

print(f"R-squared (Koeficijent determinacije): {model.rsquared:.3f}")
print(f"Prob (F-statistic) - Značajnost modela: {model.f_pvalue:.3f}\n")

print("TABELA REGRESIONIH KOEFICIJENATA I NJIHOVE ZNAČAJNOSTI:")
display(tabela.style
    .hide(axis='index')
    .set_properties(**{'text-align': 'left', 'padding': '10px', 'border-bottom': '1px solid #ddd'})
    .set_table_styles([{'selector': 'th', 'props': [('background-color', '#2E75B6'),
                                                     ('color', 'white'),
                                                     ('padding', '10px')]}])
)
```

Konstanta (const = 40.59): Predstavlja teoretski broj sati učenja za studenta koji ima GPA 0 i koji ne radi. Ovaj parametar služi za pozicioniranje regresione linije.

GPA (Coef. = -0.8589, P>|t| = 0.5604): Koeficijent pokazuje da se povećanjem prosjeka (GPA) za 1 ocjenu, broj sati učenja u prosjeku smanjuje za oko 0.86 sati. Međutim, p-vrijednost iznosi 0.5604, što je daleko veće od standardne granice značajnosti ($\alpha = 0.05$). To znači da GPA nije statistički značajan prediktor za broj sati učenja u ovom uzorku.

Zaposlen_Dummy (Coef. = -5.3273, P>|t| = 0.3020): Student koji radi uz studij u prosjeku uči oko 5.33 sati manje sedmično u odnosu na studenta koji ne radi. Ipak, p-vrijednost iznosi 0.3020 (veća od 0.05), što znači da ni radni status nije statistički značajan prediktor u ovom modelu.

Vrijednost $R^2$ iznosi 0.011 (odnosno 1.1%), dok je prilagođeni $R^2$ (Adj. R-squared) čak blago negativan (-0.005). Ovo nam govori da izabrani prediktori (GPA i radni status) objašnjavaju samo 1.1% varijabiliteta u broju sati učenja studenata. Preostalih 98.9% varijabiliteta uzrokovano je drugim faktorima koje nismo uključili u model.

Prob (F-statistic) iznosi 0.505, što je znatno veće od 0.05. Ovo ukazuje na to da cijeli regresioni model uopšteno nije statistički značajan, odnosno kombinacija GPA i zaposlenja ne predviđa vrijeme učenja bolje od obične aritmetičke sredine.
